```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('mod.csv')
df
```

```python
df1 = df.query('result == 0')
```

```python
plt.figure(figsize = (12,7))

plt.scatter(data = df, y = 'result', x = 'temp', alpha = 0.7, s = 150, linewidths = 1, color = 'red')
plt.scatter(data = df1, y = 'result', x = 'temp', alpha = 0.7, s = 150, linewidths = 1, color = 'blue')
plt.xlim(20,90)

plt.title('Defect of Space Shuttle O-rings (Chart: BUS9551)');
plt.xlabel('Outside temperature(℉)')
plt.ylabel('Total O-Ring Damage')
plt.legend(['Damaged', 'Passed'])

plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```
